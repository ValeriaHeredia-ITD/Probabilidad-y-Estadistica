```python
#Determinar mediante la regla de sturges la cantidad optima de bins o separadores
import math

df = pd.DataFrame(carros)
df
```

```python
n = len(carros)
K = int(1 + math.log2(n))
print("Valor de n =", n, " \nNumero de intervalos =", K)
```

```python
#Crear la tabla de frecuencia
tabla_frecuencia = pd.cut(df['price'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ["Rango de Precios", "Frecuencia"]
tabla_frecuencia
```

```python
#Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Rango de Precios')
tabla_frecuencia
```

```python
import matplotlib.pyplot as plt

#Crear un histograma basado en la tabla de frecuencia
plt.hist(df['price'], bins=K, alpha=0.5)
plt.title("Histograma de edades basado en la regla de Sturges")
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```



---

### **Interpretación y conclusión**

---

En la tabla se distribuyen los precios de 205 carros, se dividieron en 8 intervalos usando la regla de Sturges.

La mayoría de los carros (98) tienen un precio relativamente bajo al estar en el primer rango de precios, entonces la tabla está dominada por carros de bajo precio.

De ahí va disminuyendo la cantidad de frecuencia en los precios de los autos, es decir, entre mayor sea el precio menos son los carros con esos costos (existen más baratos que caros).

```python
#Crear un diagrama de cajas para visualizar la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)
sns.boxplot(x="price", y="body-style", data=carros)
```



---

### **Interpretación y conclusión**

---

En este diagrama de cajas y bigotes muestra cómo se distribuyen los precios de los carros según su tipo de carrocería.

Por ejemplo, los convertibles tienen la mediana de precio más alto y distribución en sus precios variada, y los hardtops tienen una gran dispersión en sus precios.
Los valores atípicos, que son los circulitos, se muestran cuando algún modelo es más caro que el resto.

Con esto puedo deducir que eel tipo de carrocería influye en el precio del carro, al parecer los convertibles y hardtops son más caros. También que los valores atípicos pueden ser algún modelo premium.

```python
#Crear un diagrama de cajas para visualizar la relación estadística entre la columna del precio del automóvil (price) y tipo tracción ruedas (drive wheels)
sns.boxplot(x="price", y="drive-wheels", data=carros)
```



---

### **Interpretación y conclusión**

---

Este diagrama se distribuye por el precio y el tipo de tracción en las ruedas, donde rwd es tracción trasera, fwd es tracción delantera y 4wd es tracción en las cuatro ruedas.

En rwd presenta precios más elevados, su dispersión de precios es más grande y también hay varios valores atípicos que indican que algunos carros rwd son mucho más caros al promedio.
La fwd tiene menos dispersión en sus precios, es decir, se mantiene en cierto rango un poco más económico. También cuenta con algunos valores atípicos.
Por último el 4wd, es el que menos datos tiene, sus precios están entre 8,000 y 18,000 y cuenta con menos valores atípicos a los anteriores.

Por lo que se puede notar, la tracción trasera suele ser más costosa, pero a su vez varian sus precios, por lo que puede incluir modelos más económicos o de gama alta.
