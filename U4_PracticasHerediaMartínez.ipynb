## **Ejercicio 1.**
## Distribución Normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Generar datos para la distribución normal
media = 70 #Media
desv_std = 4 #Desviacion estrandar
num_datos = 500

#Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
#Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

#Calcular la PDF de la distribucion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)** 2 / (2 * desv_std **2) )
```

```python
print("Imprimir los primeros 50 valores de x", x[0:50], " , y los primeros 50 pdf: ", pdf[0:50])
```

```python
#Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b', density=True)
plt.plot(x, pdf, 'k', linewidth=2)

plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

# **- La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.**

---



```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

media = 70
desv_std = 4
total_estudiantes = 500

peso_inferior = 68
peso_superior = 72

prob_entre_68_72 = norm.cdf(peso_superior, media, desv_std) - norm.cdf(peso_inferior, media, desv_std)
estudiantes_entre_68_72 = total_estudiantes * prob_entre_68_72

print("Probabilidad de que un estudiante pese entre 68 kg y 72 kg:", prob_entre_68_72)
print("Número de estudiantes que pesan entre 68 kg y 72 kg:", estudiantes_entre_68_72)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, linewidth=2, color='b')

plt.fill_between(x, y, where=(x >= peso_inferior) & (x <= peso_superior), color='skyblue', alpha=0.5)

plt.title('Probabilidad de que un estudiante pese entre 68 kg y 72 kg.\n $\mu=70$, $\sigma=4$, n = 500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

### **PROBLEMA 1.**

Este es un problema de distribución de Poisson con λ (lambda) = 5.
La distribución de Poisson se utiliza para modelar la cantidad de veces que ocurre un evento en un intervalo fijo de tiempo o espacio, siempre que esos eventos ocurran de forma independiente y con una tasa promedio constante. En este caso, se trata de contar cuántas personas ingresan a una unidad de cuidados intensivos en un día, y se sabe que en promedio ingresan cinco personas por día. Como no hay un número máximo de personas que puedan ingresar y se están contando ocurrencias en un tiempo determinado, este problema es una distribución de Poisson.










```python
from scipy.stats import poisson

lambda_poisson = 5

# a) Probabilidad de que ingresen exactamente 2 personas
prob_dos_personas = poisson.pmf(2, lambda_poisson)
print("a) Probabilidad de que ingresen exactamente 2 personas:", prob_dos_personas)

# b) ¿Es probable que x exceda 10?
prob_mayor_10 = 1 - poisson.cdf(10, lambda_poisson)
print("b) Probabilidad de que ingresen más de 10 personas:", prob_mayor_10)

if prob_mayor_10 > 0.05:
    print("→ Sí, es bastante probable que ingresen más de 10 personas.")
else:
    print("→ No es muy probable que ingresen más de 10 personas.")
```


### **PROBLEMA 2.**


Se resuelve con una distribución binomial porque cuenta cuántos motores defectuosos hay en un lote de 1,000, y cada motor tiene solo dos opciones: ser defectuoso o no. Pero como el número de motores es muy grande (1,000) y la probabilidad de que un motor sea defectuoso es muy pequeña (0.001), se usa la distribución de Poisson y facilita los cálculos, se utiliza con un valor promedio (λ) igual a 𝑛 ⋅ 𝑝, que en este caso es 1.

```python
from scipy.stats import poisson

n = 1000
p = 0.001
lambda_poisson = n * p  # λ = 1

# a) Probabilidad de que ninguno sea defectuoso
prob_0_defectuosos = poisson.pmf(0, lambda_poisson)

# b) Probabilidad de que tres sean defectuosos
prob_3_defectuosos = poisson.pmf(3, lambda_poisson)

# c) Probabilidad de que cuatro sean defectuosos
prob_4_defectuosos = poisson.pmf(4, lambda_poisson)

print("a) Probabilidad de que ninguno sea defectuoso:", prob_0_defectuosos)
print("b) Probabilidad de que tres sean defectuosos:", prob_3_defectuosos)
print("c) Probabilidad de que cuatro sean defectuosos:", prob_4_defectuosos)
```
