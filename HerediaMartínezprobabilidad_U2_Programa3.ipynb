```python
import math as mt
from itertools import permutations

# NÃºmero total de asistentes
n = 10

# NÃºmero de preguntas/asistentes a seleccionar
r = 4

# Calcular el nÃºmero de permutaciones
Pr = mt.factorial(n) / mt.factorial(n - r)

print("Existen", Pr, "maneras de elegir 4 asistentes de un total de 10.")

# Generar las permutaciones
asistentes = ["Asistente " + str(i + 1) for i in range(n)]
permutaciones_asistentes = list(permutations(asistentes, r))

# Imprimir las permutaciones
print("AquÃ­ estÃ¡n las permutaciones:")
for permutacion in permutaciones_asistentes:
    print(permutacion)
```
